# Exploratory Data Analysis


### Data loading & Processing 

```python
%matplotlib inline

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)


df_read = pd.read_csv("sensor_data_sample.csv",index_col=1)  
df_read.index = pd.to_datetime(df_read.index)

df_read.head()
```

```python

df_read.index 
```

```python
df = df_read.loc['2016-01':'2016-12']
```

```python
df.index
```

```python
list(set(list(df['Station Name'])))

```

```python
df[['Measurement Timestamp Label', 'Station Name', 'Air Temperature']]
```

```python

sns.set()

#plt.plot(df.index, df[['Air Temperature', 'Wet Bulb Temperature', 'Humidity', 'Rain Intensity']])
#plt.legend(['Air Temperature', 'Wet Bulb Temperature', 'Humidity', 'Rain Intensity'], ncol=2, loc='upper left');

ax = df.loc[df['Station Name'] == list(set(list(df['Station Name'])))[0], 'Air Temperature'].plot(figsize=(20, 15), marker = 'o', color = 'b')
ax = df.loc[df['Station Name'] == list(set(list(df['Station Name'])))[1], 'Air Temperature'].plot(figsize=(20, 15), marker = 'o', color = 'r')
ax = df.loc[df['Station Name'] == list(set(list(df['Station Name'])))[2], 'Air Temperature'].plot(figsize=(20, 15), marker = 'o', color = 'g')

plt.legend([list(set(list(df['Station Name'])))[0], list(set(list(df['Station Name'])))[1], list(set(list(df['Station Name'])))[2]], ncol=1, loc='upper left', fontsize = 20);

plt.show()

```

### FaceGrid Histogram

```python
import numpy as np
bins = np.arange(0, 65, 5)
g = sns.FacetGrid(df, col='Station Name')
g = g.map(plt.hist, 'Air Temperature', bins=bins, color="r")
```

### Correlation

```python
corr = df.corr(method = 'pearson')
```

```python
fig, ax = plt.subplots(figsize=(10,10))   

ax = sns.heatmap(corr, annot=True, cmap="YlGnBu", linewidths=.2, fmt='.2f')
```
